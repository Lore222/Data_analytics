```python
import json
```

```python
cmd = None

def load_data():
    try:
        with open('dati_progetto.json', 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        data = {"product": [], "sale": []}
        save_data(data)
        return data

def save_data(data):
    with open('dati_progetto.json', 'w') as file:
        json.dump(data, file)

data = load_data()
warehouse = {item['nome']: item for item in data['product']}

def save_data_to_file():
    data['product'] = list(warehouse.values())
    save_data(data)
```

```python
while True:
    cmd = input("Inserisci un comando: ")

    if cmd == "vendita":
        product_name = input("Nome del prodotto: ")

        while True:
            try:
                quantity = float(input("Quantit√†: "))
                if quantity > 0:
                    break
                else:
                    print("Errore: Inserisci un numero positivo.")
            except ValueError:
                print("Errore: Inserisci un numero valido.")

        if product_name in warehouse:
            product = warehouse[product_name]
            if product["quantity"] >= quantity:
                sale_amount = quantity * product["selling_price"]
                product["quantity"] -= quantity
                save_data_to_file()
                print(f"VENDITA REGISTRATA: {quantity} X {product_name}: ‚Ç¨{sale_amount:.2f}")
            else:
                print("Quantit√† insufficienti nel magazzino.")
        else:
            print("Prodotto non trovato nel magazzino.")

    elif cmd == "aggiungi":
        product_name = input("Nome del prodotto: ")

        while True:
            try:
                quantity = float(input("Quantit√†: "))
                if quantity > 0:
                    break
                else:
                    print("Errore: Inserisci un numero positivo.")
            except ValueError:
                print("Errore: Inserisci un numero valido.")

        purchase_price = float(input("Prezzo di acquisto: Euro "))
        selling_price = float(input("Prezzo di vendita: Euro "))

        if product_name in warehouse:
            product = warehouse[product_name]
            product["quantity"] += quantity
            save_data_to_file()
            print(f"AGGIORNATO: {quantity} X {product_name}")
        else:
            warehouse[product_name] = {"nome": product_name, "quantity": quantity, "purchase_price": purchase_price, "selling_price": selling_price}
            save_data_to_file()
            print(f"AGGIUNTO: {quantity} X {product_name}")

    elif cmd == "profitti":
        gross_profit = sum(product["quantity"] * product["selling_price"] for product in warehouse.values())
        net_profit = sum(product["quantity"] * (product["selling_price"] - product["purchase_price"]) for product in warehouse.values())
        print(f"Profitto: lordo=‚Ç¨ {gross_profit:.2f} netto=‚Ç¨ {net_profit:.2f}")

    elif cmd == "elenca":
        print("PRODOTTO QUANTITA' PREZZO")
        for product in warehouse.values():
            print(f"{product['nome']} {product['quantity']} ‚Ç¨ {product['selling_price']:.2f}")


    elif cmd == "aiuto":
        print("I comandi disponibili sono i seguenti:")
        print("‚Ä¢ aggiungi: aggiungi un prodotto al magazzino")
        print("‚Ä¢ elenca: elenca i prodotti in magazzino")
        print("‚Ä¢ vendita: registra una vendita effettuata")
        print("‚Ä¢ profitti: mostra i profitti totali")
        print("‚Ä¢ aiuto: mostra i possibili comandi")
        print("‚Ä¢ chiudi: esci dal programma")

    elif cmd == "chiudi":
        print("Bye bye.. üôè GRAZIE")
        break  

    else:
        print("Comando non valido. Usa 'aiuto' per vedere i comandi disponibili.")
```
